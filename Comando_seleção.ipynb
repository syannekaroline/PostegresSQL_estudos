### 1. Comando de Seleção

#### Permite selecionar os dados dentro de uma tabela

**SELECT**

#### Sintaxe
* Opção 1:

**SELECT coluna1, coluna2 FROM tabela; -> usar o comando select e selecionar as colunas específicas de uma tabela**

* Opção 2:


**SELECT * FROM tabela; --> Seleciona todas as colunas de uma tabela**


```python
import  os 
import  psycopg2  as  ps 
import  pandas  as  pd

conn = ps.connect(
    host=os.environ["db_host"],
    port=os.environ["db_port"],
    dbname=os.environ["db_name"],
    user=os.environ["db_user"],
    password=os.environ["db_password"])
```

```python
df = pd.read_sql_query("select * from products", con=conn)
df
```

```python
# Consulta 1: Selecione a tabela de produtos 
# SELECT * FROM products;

df = pd.read_sql_query("SELECT * FROM products", con=conn)
df
```

```python
''' Consulta 2: Selecione a tabela de clientes 

SELECT * FROM customers;'''

df = pd.read_sql_query("SELECT * FROM customers", con=conn)
df
```

```python
'''# Consulta 3: Selecione a tabela de funcionários

SELECT * FROM employees;
'''
df = pd.read_sql_query("SELECT * FROM employees", con=conn)
df
```

```python
'''Consulta 4: Selecione as colunas: employee_id, first_name, last_name e address da tabela funcionários

SELECT employee_id, first_name, last_name, address FROM employees''' 
df = pd.read_sql_query("SELECT employee_id, first_name, last_name, address FROM employees", con=conn)
df
```
